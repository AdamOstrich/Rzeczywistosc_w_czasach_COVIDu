```julia
using DelimitedFiles
zarazenia=readdlm("dane_zarazen_COVID.txt")
ccc_covid=readdlm("CCC_COVID.txt")
ccc_przed=readdlm("CCC_PRZED.txt")
cd_prj_covid=readdlm("CD_PROJEKT_COVID.txt")
cd_prj_przed=readdlm("CD_PROJEKT_PRZED.txt")
kghm_covid=readdlm("KGHM_COVID.txt")
kghm_przed=readdlm("KGHM_PRZED.txt")
orlen_covid=readdlm("PKNORLEN_COVID.txt")
orlen_przed=readdlm("PKNORLEN_PRZED.txt")
sant_covid=readdlm("SANTANDER_COVID.txt")
sant_przed=readdlm("SANTANDER_PRZED.txt")
zaraz_bez_weekend=readdlm("zarazeni_w_dni_GPW.txt")
powietrze_covid=readdlm("powietrze_covid.txt")
powietrze_przed=readdlm("powietrze_przed.txt")
```

```julia
zarazenia_plus_weekendy = Float16.(zarazenia[2:end, 3])
ccc_covid_wolumen=Float16.(ccc_covid[2:end, 6])
ccc_covid_otwarcie=Float16.(ccc_covid[2:end, 2])
ccc_przed_wolumen=Float16.(ccc_przed[2:end, 6])
ccc_przed_otwarcie=Float16.(ccc_przed[2:end, 2])
cd_covid_wolumen=Float16.(cd_prj_covid[2:end, 6])
cd_covid_otwarcie=Float16.(cd_prj_covid[2:end, 2])
cd_przed_wolumen=Float16.(cd_prj_przed[2:end, 6])
cd_przed_otwarcie=Float16.(cd_prj_przed[2:end, 2])
kghm_covid_wolumen=Float16.(kghm_covid[2:end, 6])
kghm_covid_otwarcie=Float16.(kghm_covid[2:end, 2])
kghm_przed_wolumen=Float16.(kghm_przed[2:end, 6])
kghm_przed_otwarcie=Float16.(kghm_przed[2:end, 2])
orlen_covid_wolumen=Float16.(orlen_covid[2:end, 6])
orlen_covid_otwarcie=Float16.(orlen_covid[2:end, 2])
orlen_przed_wolumen=Float16.(orlen_przed[2:end, 6])
orlen_przed_otwarcie=Float16.(orlen_przed[2:end, 2])
sant_covid_wolumen=Float16.(sant_covid[2:end, 6])
sant_covid_otwarcie=Float16.(sant_covid[2:end, 2])
sant_przed_wolumen=Float16.(sant_przed[2:end, 6])
sant_przed_otwarcie=Float16.(sant_przed[2:end, 2])
zaraz_bezweekend=Float16.(zaraz_bez_weekend[2:end, 3])
```
